```python
import numpy as np   
import pandas as pd  #-> to create panda dataframe, from csv to well structured dataframe to understand the data well

import difflib # since we need to take input from user, user can make spelling mistake, we need to find the close correct possible name by user by comparing with the movie data.

from sklearn.feature_extraction.text import TfidfVectorizer # this helps to convert textual data to numerical value(Feature vectors), we have title of movie,
from sklearn.metrics.pairwise import cosine_similarity # gives similarity score for all diff movie compare to other movies. user will give some movie name so we will find the similarity of that with all other movies and we have to find which move have highest similarity score.
```

Data collection and pre-process

```python
movies_data=pd.read_csv(r"D:\ML_WORK\movies.csv")
```

```python
movies_data.head()
```

```python
movies_data.shape
```

```python

selected_features=["genres",'keywords','tagline','cast','director']
print(selected_features)
```

```python
#replacing null values with null string
for feature in selected_features:
    movies_data[feature]=movies_data[feature].fillna('')
```

```python
#combining all the features(concating all the features all together)

combined_features=movies_data['genres']+" "+movies_data['keywords']+" "+movies_data['tagline']+" "+movies_data['cast']+" "+movies_data['director']
```

```python
combined_features.head
```

```python
#After combining all features, now converting this textual data to feature vector( numeric value)
#we are going to use tfidf to convert this, so than we can easily calculate the cosine value with numeric values
#creating instance of vectorizer
vectorizer=TfidfVectorizer() #-->loading this TfidVec.. that we have imported from Sklearn into vectorizer
```

```python
#now fitting and transforming all the data
feature_vectors=vectorizer.fit_transform(combined_features)
```

```python
print(feature_vectors) # now we can fit this data into the ML model this will give numeric value to text data
```

```python
#Now finding the similarity score-> using cosine similarity, also imported from sklearn

#we will get diff. score value for all movie, so we will store this in list
```

```python
similarity=cosine_similarity(feature_vectors)
```

```python
print(similarity)
```

```python
print(similarity[0][:9]) # so basically it will pick 1st movie, and compare with all to give a similarity score, similarly for all movie.
```

```python
similarity.shape #1. is movie index, 2nd is similarity score
```

```python
# now we can ask user to give input fav movie name, 
# than will check if that movie name is present in dataset, than will return the similar movie based on similarity score

movie_name=input("Enter your favourite movie name : ")
```

```python
#create a list that contain all the movie names givenin that dataset
# so that we can compare it with the given name

list_of_all_titles=movies_data['title'].tolist()
print(list_of_all_titles)
```

```python
#finding the close match for the movie name given by user

find_close_match= difflib.get_close_matches(movie_name,list_of_all_titles)
print(find_close_match)
```

```python
# we want only 1 value, which will match exactly, not all, only the most similar

close_match= find_close_match[0]
print(close_match)
```

```python
#finding the index of this movie with title, based on title

index_of_the_movie=movies_data[movies_data.title==close_match]['index'].values[0]
print(index_of_the_movie)
```

```python
#getting list of similar movie->now we will take this index and find the similar movie to this

similarity_score= list(enumerate(similarity[index_of_the_movie]))  #enumerate is used to carry a loop in list

print(similarity_score) # 1st value index of movie, 2nd is similarity score compare to input
```

```python
len(similarity_score)
```

```python
# we want only which has highest similarity value, we will sort this from high similarity value to lowest

sorted_similar_movies=sorted(similarity_score,key= lambda x:x[1],reverse=True)  #key= lambda x:x[1]--> this means simlarity score value(2nd part). means the order will be based on score value not on index based
print(sorted_similar_movies)
```

```python
#we will print the name of similar movie based on index

print('Movies suggested for you :\n')

i=1
for movie in sorted_similar_movies:
    index=movie[0]
    title_from_index=movies_data[movies_data.index==index]['title'].values[0]
    if (i<16):
        print(i,'.',title_from_index)
        i+=1
```
